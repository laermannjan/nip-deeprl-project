```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

# REWARD  LL_150_150_150_rewards

```python
mean_rew = []
for i in range(3):
    mean_rew += [np.load('{}LL_150-150-150_rewards.npy'.format(i+1))]
mean_rew = np.array(mean_rew)
mean_rewC = np.copy(mean_rew)
print (mean_rew.shape)
div = np.copy(np.max(mean_rew[0]))
for i in range(3):
    print ((mean_rew[i].shape))
    #mean_rew[i] = 150*mean_rew[i][:]/div #mean_rew[0].shape[0]
for i in range(3):
    print (mean_rew[i].shape)

# LL_150_150_150_meanrewards

fig = plt.figure(figsize=(10,10))
fig.suptitle("LL_150_150_150_rewards", fontsize=17)
plt.subplot2grid((3, 3), (1, 0))
plt.plot(range(mean_rew[0].shape[0]),mean_rew[0],linewidth=.2)
plt.xlabel('episodes')
plt.title('reward 1st trial')
plt.ylim(-450,350)
#plt.xlim(0,mean_rew.mean(0).shape[0])
plt.grid()
plt.subplot2grid((3, 3), (1, 1))
plt.plot(range(mean_rew[1].shape[0]),mean_rew[1],linewidth=.2)
plt.xlabel('episodes')
plt.title('reward 2nd trial')
plt.ylim(-450,350)
#plt.xlim(0,mean_rew.mean(0).shape[0])
plt.grid()
plt.subplot2grid((3, 3), (1, 2))
plt.plot(range(mean_rew[2].shape[0]),mean_rew[2],linewidth=.2)
plt.xlabel('episodes')
plt.title('reward 3rd trial')
plt.ylim(-450,350)
#plt.xlim(0,mean_rew.mean(0).shape[0])
plt.grid()
plt.subplot2grid((3, 3), (0, 0), colspan=3)
for i in range(3):
    #print ((mean_rew[i].shape))
    mean_rewC[i] = 150*mean_rewC[i][:mean_rew[0].shape[0]]/div #
plt.plot(range(mean_rewC.mean(0).shape[0]),mean_rewC.mean(0),linewidth=.2)

#plt.xlabel('episodes')
plt.ylabel('avg reward over 3 trials')
plt.ylim(-250,250)
plt.xlim(0,mean_rewC.mean(0).shape[0])
plt.grid()
plt.subplot2grid((3, 3), (2, 0), colspan=3)
for i in range(3):
    plt.plot(range(mean_rew[0].shape[0]),mean_rew[i][:mean_rew[0].shape[0]],linewidth=.2,alpha=.7)
plt.ylabel('reward over 3 trials')
plt.ylim(-250,250)
plt.xlim(0,mean_rewC.mean(0).shape[0])
plt.grid()    
```
