# Influence of different parameters on our DQN


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
rwd = []
avg_rwd = []
stp = []
avg_stp = []
lss = []
for j in range (3):  # loop over 5 different hyper params
    reward_array = []
    step_array = []
    loss_array = []
    avg_rwd_array = []
    avg_stp_array = []
    for i in range(10):
        reward = np.load('log/train/{}_train_ERL_{}.npy'.format(j+1,i))
        reward_array += [reward]
        step = np.load('log/train/{}_train_ESL_{}.npy'.format(j+1,i))
        step_array += [step]
        loss = np.load('log/train/{}_train_L_{}.npy'.format(j+1,i))
        loss_array += [loss]
        for k in range(len(reward_array[i])-99):
            avg_rwd_array.append(np.average(reward_array[i][k:k+100]))
        for m in range(len(step_array[i])-99):
            avg_stp_array.append(np.average(step_array[i][m:m+100]))
        
    reward_array = np.array(reward_array)
    step_array = np.array(step_array)
    loss_array = np.array(loss_array)
    avg_rwd_array = np.array(avg_rwd_array)
    avg_stp_array = np.array(avg_stp_array)
    rwd  += [reward_array.T]
    stp  += [step_array.T]
    lss  += [loss_array.T]
    avg_rwd += [avg_rwd_array]
    avg_stp += [avg_stp_array]
    
rwd = np.array(rwd).T
stp = np.array(stp).T
lss = np.array(lss).T
avg_rwd  = np.array(avg_rwd ).T
#print(avg_rwd)
avg_stp = np.array(avg_stp ).T
```

```python
sns.tsplot(data=rwd)
plt.ylabel('reward')
plt.xlabel('episodes')
```

```python
sns.tsplot(data=avg_rwd)
plt.ylabel('reward')
plt.xlabel('average reward per 100 episodes')
```

```python
sns.tsplot(data=stp)
plt.ylabel('steps')
plt.xlabel('episodes')
```

```python
sns.tsplot(data=avg_stp)
plt.ylabel('reward')
plt.xlabel('average number of steps per 100 episodes')
```

```python
sns.tsplot(data=lss)
plt.ylabel('loss')
plt.xlabel('epochs')
```
